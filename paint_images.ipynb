```python
# 模拟信息度量的大小
import numpy as np
import matplotlib
import scienceplots
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np 
plt.style.use('science')
#font = {'family' : 'Times New Roman',
#        'weight' : 'regular',
#        'size'   : 14}
#matplotlib.rc('font', **font)

# 生成5个二维高斯分布N
mus = np.array([[0, 0],
                [3, 3],
                [3,-3],
                [-3, 3],
                [-3, -3]])  # 均值
sigmas = np.array([[[1, 0], [0, 1]],
                   [[2, 0.5], [0.5, 1]],
                   [[1, 0.5], [0.5, 1]],
                   [[1, 0.5], [0.5, 1]],
                   [[1, 0.5], [0.5, 1]]])  # 协方差矩阵

Ns = []
for i in range(5):
    Ns.append(np.random.multivariate_normal(mus[i], sigmas[i], size=1000))
# 绘制散布图
plt.scatter(Ns[0][:, 0], Ns[0][:, 1], marker="o", s=7, alpha=0.5)
plt.scatter(Ns[1][:, 0], Ns[1][:, 1], marker="v", s=7, alpha=0.5)
plt.scatter(Ns[2][:, 0], Ns[2][:, 1], marker="+", s=7, alpha=0.5)
plt.scatter(Ns[3][:, 0], Ns[3][:, 1], marker="d", s=7, alpha=0.5)
plt.scatter(Ns[4][:, 0], Ns[4][:, 1], marker="x", s=7, alpha=0.5)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Gaussian Distribution')
# 显示图像
plt.show()
#plt.savefig("/home/wyx/vscode_projects/SSAL/test/simulation.pdf", dpi=300)

```

```python
# 绘制样本选择的图形
import json,copy
import os,sys
#sys.path.append(os.path.dirname(os.path.dirname(__file__)))
import numpy as np
from PIL import Image
import matplotlib
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from CIFAR10.dataset import CIFAR10DatasetAL
from FashionMNIST.dataset import FashionMNISTDatasetAL
import seaborn as sns
palette = sns.color_palette('hls', 10)
np.random.seed(5)

# font1 = {'family' : 'Times New Roman',
# 'weight' : 'normal',
# 'size'   : 18,
# }
#plt.style.use('science')
#matplotlib.rc('font', **font1)
sys.path.append(os.path.dirname(os.path.dirname(os.getcwd())))

dataset_name = "CIFAR10"
info_metric = "F"
balance = "true"

queryid = 10

# 使用Logit作为可视化的结果
data_path = "/home/wyx/vscode_projects/SSAL/models/{}/{}-{}/logits.{}.json".format(dataset_name, info_metric, balance,queryid)

state_path = "/home/wyx/vscode_projects/SSAL/models/{}/{}-{}/states.json".format(dataset_name, info_metric, balance)

dataset_L, dataset_U = CIFAR10DatasetAL.createLURandomly(1000)

data = json.load(open(data_path, "r"))
state = json.load(open(state_path, "r"))

pre_labels = {}
for idx in data:
    pre_labels[idx] = np.argmax(data[idx])

gt_labels = {}
for idx in data:
    gt_labels[idx] = dataset_U.data[idx][0]

# 由于样本量太多，因此随机挑选一部分进行可视化
# 不包含所选择的样本
n_vis = 5000

select_idx = state[queryid-1]["select_ids"]
s_idx_set = set(select_idx)
data_idx = list(data)
np.random.shuffle(data_idx)
vis_idx = []
count = 0
for id in data_idx:
    if id in s_idx_set:
        continue
    vis_idx.append(id)
    count += 1
    if count >= n_vis:
        break

# 合并未选择样本和选择的样本
vis_idx.extend(copy.deepcopy(select_idx))
print(len(vis_idx))
print(len(select_idx))
assert len(vis_idx) == len(set(vis_idx))

vis_idx = [int(x) for x in list(vis_idx)]
vis_idx = sorted(vis_idx)

d = len(data[str(data_idx[0])])
n = len(vis_idx)
vis_mat = np.empty((n,d)).astype(np.float16)
itoid_mapping = {}
idtoi_mapping = {}
for i, id in enumerate(vis_idx):
    vis_mat[i] = np.asarray(data[str(id)]).astype(np.float16)
    itoid_mapping[i] = str(id)
    idtoi_mapping[str(id)] = i

X_tsne = TSNE(n_components=2, random_state=33).fit_transform(vis_mat)
```

```python

select_idx_mapping = []
for idx in select_idx:
    cc_idx = vis_idx.index(int(idx))
    select_idx_mapping.append(cc_idx)

# Draw
plt.figure(figsize=(15, 15))
for i in range(n):
    #c = pre_labels[itoid_mapping[i]]
    c = gt_labels[itoid_mapping[i]]
    if itoid_mapping[i] not in s_idx_set:
        plt.scatter(X_tsne[i, 0], X_tsne[i, 1], color=palette[c])

plt.scatter(X_tsne[select_idx_mapping, 0], X_tsne[select_idx_mapping, 1], marker="*", color="red", label="Selected Samples")
plt.xticks([])
plt.yticks([])
#plt.legend(fancybox=True,ncol=1,bbox_to_anchor=(0.7, 0.1),prop = {'size':20})
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.savefig("/home/wyx/vscode_projects/SSAL/models/{}/{}-{}/{}-tsne.pdf".format(dataset_name,info_metric,balance,dataset_name), dpi=300)
plt.show()
```
